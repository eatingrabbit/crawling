```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
# 1) 브라우저 열기
driver = webdriver.Chrome() #크롬 열기
url = "https://brand.naver.com/aestura365/products/8157827580?is_retargeting=true&c=260101_p_Naver_product&pid=Naver&deep_link_value=https%3A%2F%2Fbrand.naver.com%2Faestura365%2Fproducts%2F8157827580%3Fka_ref%3Dkakao%26ka_req_id%3DgEUKsUVYwDyYJLaY50qEAA%26ka_template%3Dcommerce_basic&dtm_source=KK&dtm_detail=gEUKsUVYwDyYJLaY50qEAA&dtm_campaign=commerce_basic&dtm_medium=OG"
driver.get(url) #사이트 접속 
```

```python
# 리뷰 버튼 클릭
review_button = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[4]/ul/li[2]/a")
review_button.click()
time.sleep(2)
# 최신순 클릭
latest_button = driver.find_element(By.XPATH, '//*[@id="REVIEW"]/div/div[3]/div[1]/div[1]/ul/li[2]/a')
latest_button.click()
```

```python
# 리뷰 개수
review_count_text = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[7]/div/div[3]/div[1]/div[1]/strong/span").text
review_count = int(review_count_text.replace(",", ""))
print(review_count)
```

```python
# 리뷰 리스트
# final_list = []
while len(final_list) < 2500:
    try:
        review_elem_list = driver.find_elements(By.XPATH, "/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[7]/div/div[3]/div[2]/ul/li/div/div/div/div[1]/div/div[1]")
        
        # 현재 페이지 리뷰 추출
        for review_elem in review_elem_list:
            star = review_elem.find_element(By.XPATH, './/div[1]/div[2]/div[1]/em')
            date = review_elem.find_element(By.XPATH, './/div[1]/div[2]/div[2]/span')
            text_span = review_elem.find_elements(By.XPATH, './div[2]/div/span')
            if len(text_span) == 1:
                text = text_span[0]
            elif len(text_span) >= 2:
                text = text_span[-1]
            # print(star.text, date.text, text.text)
            final_list.append((star.text, date.text, text.text))
        
        # 다음 페이지로 이동
        page_button_div = driver.find_element(By.XPATH, '//*[@id="REVIEW"]/div/div[3]/div[2]/div/div')
        current_page = int(page_button_div.find_element(By.CSS_SELECTOR, 'a[aria-current="true"]').text)
        if current_page % 10 == 0: # "다음" 버튼 클릭
            next_button = page_button_div.find_elements(By.TAG_NAME, "a")[-1]
            next_button.click()
        else: # 현재 페이지 + 1 버튼 클릭
            next_button = page_button_div.find_element(By.XPATH, f".//a[text()='{current_page + 1}']")
            next_button.click()
        time.sleep(1) # 1초 대기 (stale 에러 방지)
        
        print(f"완료된 현재 페이지 번호: {current_page}")
        print("현재까지 수집된 리뷰 수: ", len(final_list))
        
    except Exception as e:
        print(e)
        break
print("현재까지 수집된 리뷰 수: ", len(final_list))
```

```python
df = pd.DataFrame(final_list, columns=['별점', '날짜', '리뷰내용'])
df.to_excel('naver_review_result.xlsx', index=False)
print(f"총 {len(df)}개 리뷰 저장 완료")
```
