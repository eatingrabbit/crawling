```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
# 1) 브라우저 열기
driver = webdriver.Chrome() #크롬 열기
url = "https://mudit.kr/product/%EB%AE%A4%EB%94%94%ED%8A%B8-%EB%B8%8C%EB%A1%9C%EB%A9%9C%EB%9D%BC%EC%9D%B8/15/"
driver.get(url) #사이트 접속 
```

```python
# 리뷰 버튼 클릭
review_button = driver.find_element(By.XPATH, '//*[@id="prdDetail"]/div[2]/ul/li[3]/a')
review_button.click()
time.sleep(2)
```

```python
# 최신순 클릭
# iframe이 내부에 삽입되어 있음 => driver를 해당 iframe 내부로 들여보내기
iframe = driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div/div/div[4]/div[4]/div[2]/iframe')
driver.switch_to.frame(iframe)
# button.click으로 물리적 클릭 대신 JS 스크립트 실행으로 엘리먼트에 직접 클릭 이벤트 실행
latest_button = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[2]/div[7]/div/div[1]/div/div[2]/span')
driver.execute_script("arguments[0].click();", latest_button)
```

```python
# # 리뷰 개수
# review_count_text = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[7]/div/div[3]/div[1]/div[1]/strong/span").text
# review_count = int(review_count_text.replace(",", ""))
# print(review_count)
```

```python
# 리뷰 리스트
final_list = []
while len(final_list) < 2500:
    try:
        review_elem_list = driver.find_elements(By.XPATH, "/html/body/div/div[1]/div[2]/div[8]/div")
        
        # 현재 페이지 리뷰 추출
        for review_elem in review_elem_list:
            star_span = review_elem.find_element(By.XPATH, './div[1]/div[2]/div[1]/span')
            text_span = review_elem.find_element(By.XPATH, './div[1]/div[3]/div[1]/span')
            inner_html = text_span.get_attribute('innerHTML')
            
            # 내부 html 파싱
            soup = BeautifulSoup(inner_html, 'html.parser')
            # 신고하기 버튼 제거
            report_div = soup.find('div', class_='review_report_area') 
            if report_div:
                report_div.decompose()
            # 네이버페이 구매평 문구 제거
            naver_pay_spans = soup.find_all('span', string=re.compile("네이버 페이 구매평"))
            for span in naver_pay_spans:
                span.decompose()
            final_review_text = soup.get_text(separator='\n', strip=True)

            # print(final_review_text)
            final_list.append((star_span.text, final_review_text))
        
        # 다음 페이지로 이동
        css_path = "#pagination img.pagenation_btn_right"

        next_button_img = driver.find_element(By.CSS_SELECTOR, css_path)
        driver.execute_script("arguments[0].click();", next_button_img)
        time.sleep(1) # 1초 대기 (stale 에러 방지)
        
        print("현재까지 수집된 리뷰 수: ", len(final_list))
        
    except Exception as e:
        print(e)
        break
print("현재까지 수집된 리뷰 수: ", len(final_list))
```

```python
df = pd.DataFrame(final_list, columns=['별점', '리뷰내용'])
df.to_excel('output/mudit_review_result.xlsx', index=False)
print(f"총 {len(df)}개 리뷰 저장 완료")
```
