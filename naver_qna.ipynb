```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
# 1) 브라우저 열기
driver = webdriver.Chrome() #크롬 열기
url = "https://brand.naver.com/aestura365/products/8157827580?is_retargeting=true&c=260101_p_Naver_product&pid=Naver&deep_link_value=https%3A%2F%2Fbrand.naver.com%2Faestura365%2Fproducts%2F8157827580%3Fka_ref%3Dkakao%26ka_req_id%3DgEUKsUVYwDyYJLaY50qEAA%26ka_template%3Dcommerce_basic&dtm_source=KK&dtm_detail=gEUKsUVYwDyYJLaY50qEAA&dtm_campaign=commerce_basic&dtm_medium=OG"
driver.get(url) #사이트 접속 
```

```python
# QnA 버튼 클릭
qna_button = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[4]/ul/li[3]/a')
qna_button.click()
```

```python
# 리뷰 개수
qna_count_text = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[5]/div/div[3]/ul/li[3]/a/span").text
qna_count = int(qna_count_text.replace(",", ""))
print(qna_count)
```

```python
from selenium.common import NoSuchElementException
review_elem_list = driver.find_elements(By.XPATH, "/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[8]/div/div[2]/ul/li")
        
# 현재 페이지 QnA 추출
for review_elem in review_elem_list:
    question = review_elem.find_element(By.XPATH, './/a/span/span[1]').text
    if question == '비밀글입니다.':
        continue
    is_answer_completed = review_elem.find_element(By.XPATH, './div/div').text
    if is_answer_completed == '미답변':
        continue
    expand_button = review_elem.find_element(By.XPATH, './div/div/a')
    expand_button.click()
    # time.sleep(0.2)

    try:
        answer = review_elem.find_element(By.XPATH, './div[2]/div[2]/div/p').text
    except NoSuchElementException:
        continue
    print(question)
```

```python
# 리뷰 리스트
# final_list = []
prev_page = 0
while len(final_list) < qna_count:
    try:
        review_elem_list = driver.find_elements(By.XPATH, "/html/body/div[1]/div/div[3]/div/div[2]/div/div[3]/div[8]/div/div[2]/ul/li")
        
        # 현재 페이지 QnA 추출
        cur_page_list = []
        for review_elem in review_elem_list:
            question = review_elem.find_element(By.XPATH, './/a/span/span[1]').text
            if question == '비밀글입니다.':
                continue
            is_answer_completed = review_elem.find_element(By.XPATH, './div/div').text
            if is_answer_completed == '미답변':
                continue
            expand_button = review_elem.find_element(By.XPATH, './div/div/a')
            expand_button.click()
            time.sleep(0.3)

            try:
                answer = review_elem.find_element(By.XPATH, './div[2]/div[2]/div/p').text
            except NoSuchElementException: # 답변 존재 X 경우 넘어가기
                # print("답변 없음: ", question)
                continue
            
            # print(is_answer_completed, question, answer)
            cur_page_list.append((question, answer))
        
        # 트랜젝션 - 페이지에서 하나라도 실패하면 final_list에 추가되지 않음
        final_list+=cur_page_list
        
        # 다음 페이지로 이동
        page_button_div = driver.find_element(By.XPATH, '//*[@id="QNA"]/div/div[3]')
        current_page = int(page_button_div.find_element(By.CSS_SELECTOR, 'a[aria-current="true"]').text)
        if current_page % 10 == 0: # "다음" 버튼 클릭
            next_button = page_button_div.find_elements(By.TAG_NAME, "a")[-1]
            next_button.click()
        else: # 현재 페이지 + 1 버튼 클릭
            next_button = page_button_div.find_element(By.XPATH, f".//a[text()='{current_page + 1}']")
            next_button.click()
        time.sleep(1) # 1초 대기 (stale 에러 방지)
        
        print(f"완료된 현재 페이지 번호: {current_page}")
        print("현재까지 수집된 리뷰 수: ", len(final_list))
        
    except Exception as e:
        print(e)
        break
print("현재까지 수집된 리뷰 수: ", len(final_list))
```
